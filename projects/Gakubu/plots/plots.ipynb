```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
```

```python
data = pd.read_csv('ourFramework_vs_autogluon.csv')
```

```python
plt.figure(figsize=(20, 27))
matplotlib.rcParams.update({'font.size': 24})
matplotlib.rcParams['figure.dpi'] = 500
g = sns.stripplot(data=data,
                     x='auc', y='task', hue='framework', jitter=False,
                     zorder=1, size=15, palette=['red', 'purple'])
g.set_xlim(0.6625, 1.015)
plt.legend(prop={'size': 25}, fontsize=18, loc = 'upper left')
plt.ylabel('')
plt.xlabel('ROC AUC score', fontsize=28)
plt.grid()
plt.show()
```

```python
data = pd.read_csv('allFramework.csv')
```

```python
plt.figure(figsize=(20, 27))
matplotlib.rcParams.update({'font.size': 24})
matplotlib.rcParams['figure.dpi'] = 500
sns.stripplot(data=data,
              x='auc', y='task', hue='framework',
            zorder=1, size=15, jitter = True, alpha = 0.6,
              palette=['red', 'blue', 'green', 'orange', 'purple'])
plt.legend(prop={'size': 13}, fontsize=18, loc = 'upper left')
plt.ylabel('')
plt.xlabel('ROC AUC score', fontsize=28)
plt.grid()
plt.legend() 
plt.show()
```
